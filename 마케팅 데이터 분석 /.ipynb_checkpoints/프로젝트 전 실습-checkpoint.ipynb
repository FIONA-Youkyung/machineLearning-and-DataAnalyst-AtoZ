### 광고성과 지표
    - CTR(Click Through Rate) = 클릭수 / 노출수 *100
    - CPM(Cost Per Mile) = 광고비용 / 노출수 *1000
    - CPC(Cost Per Click) = 광고비용 / 클릭수
    - CPA(Cost Per Action) = 광고비용 / 구매수

```python
imp = 10000 #노출수
clk = 100 #클릭수
conv = 10 #구매수
cost = 100000 #총 사용한 광고비
```

```python
#ctr
ctr = clk/imp *100
ctr
```

```python
#cpc
cpc = cost/clk
cpc
```

```python
#cpa
cpa = cost/conv
cpa
```

## 차트에 옵션 추가하기

```python
#그래프 크기 설정
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.show()
```

```python
#제목설정
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.title('예제')
plt.show()
```

```python
#제목의 폰트 크기
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.title('예제', fontsize =10 )
plt.show()
```

```python
#x축 이름 #y축 이름
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.title('예제', fontsize =10 )
plt.xlabel('xaxis', fontsize = 18) 
plt.ylabel('yaxis', fontsize = 19)
plt.show()
```

```python
#x 축 눈금설정
#설정할 눈금의 위치, 이름, 폰트사이즈, 각도
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.title('예제', fontsize =10 )
plt.xlabel('xaxis', fontsize = 18) 
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize = 10, rotation = 0)
plt.ylabel('yaxis', fontsize = 19)
plt.show()

```

```python
 #x축 범위설정
df.plot.bar(figsize = [10, 6]) #[x축 사이즈, y축 사이즈]
plt.title('예제', fontsize =10 )
plt.xlabel('xaxis', fontsize = 18) 
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize = 10, rotation = 0)
plt.xlim([-1, 4])
plt.ylabel('yaxis', fontsize = 19)
plt.show()
```
