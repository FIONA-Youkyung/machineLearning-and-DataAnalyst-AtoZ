# NSMC를 통한 감정분석
## 감정분석에 필요한 개념 

Reference : https://www.kdnuggets.com/2018/03/5-things-sentiment-analysis-classification.html

### 1. 감정이란?


감정이란, 심리학이나 **뇌과학, 신경과학**에서 사람의 감정이라는 것을 정의 내리는 건 매우 다양하다. 
    그러면 **텍스트 마이닝** 관점에서 '감정'은 어떻게 정의하고 있을까 . 이는 , 데이터를 만드는 사람, 분석하는 사람에 따라 달라질 수 있다. 

### 2. 감정분석이란?

* Natural Language Processing, Text Analysis, Computational Lingustics, biometrics 같은 방법을 이용하여 텍스트에 숨겨져있는 저자의 의도나 정보를 찾아내는 모든 방법들을 말한다.


* Opinion Mining, Sentiment Mining, Subjectivity Analysis 라고도 불리기도 한다.
    궁극적인 목표는 문서에 숨어져있던 중요한 기분, 상태 등을 뽑아내는것, 이를 확장해서, Opinion mind라고 한다. 
    적용점은 리뷰정보, 비지니스 모델 판단
    

* 그러면 어떻게 , 뭘 찾을 것인지가 중요해진다. 
    초반의 방법들은 텍스트의 극성(Polarity)를 찾기 위해 많이 시도되었다. 대표적인 예로 <b>긍정/부정</b>으로 나누는 케이스가 있다.
    그러나 모든 감정 분석이 이렇게 양 극으로 나누는 것만 있는 것은 아니며, 
    
    
* 감정 분석은 크게 Knowledge-based approach, Machine Learning-based approach가 있다.
    - Knowledge-based approach : 2000대 초반까지는 데이터를 많이 모아서, 그 데이터를 확실한 기법들, 알고있던 정보를 모아서 knowledge base를 만들었다. 그래서 이 시기에는 대규모 데이터를 모아서 정제하는 것이 중요했다. 
      그런데 데이터가 너무 많아지고, 개념들이 빠르게 생기기 시작하여서 knowledge base로 만드는데 한계에 부딪혔다. 
      > Knowledge-based는 알려진 어구, 어미, 관용 표현등을 활용하여 이미 문서들을 human expert가 평가한 데이터를 가져와 평가하는 방법이다.
      
    - Machine Learning-based approach : 그래서 요즘은 데이터의 패턴을 찾는 머신러닝 기법으로 많이 바뀌었다. 데이터를 주고, 분류하는 접근법이 가장 simple한 방법. 
      > ML-based approach는 supervised, unsupervised 방법이 있다. 최근 pretrained Language Model이 비약적으로 발달함에 따라 unsupervised 방법의 성능도 많이 높아졌지만, 아직까지는 성능면에선 **supervised가 월등히 높다.**

#### 따라서, 가장 많이 쓰이는 방법인 text classification으로서의 sentiment analysis를 공부해보고, 추가적으로 생각해볼 여러 가지 이슈에 대해 고민해보는 시간을 가진다.

#### 정리하면, 감정 분석이란 결국엔 텍스트 이면에 있는 저자의 의도를 파악하는 작업인데, 이 저자의 의도라는 것이 굉장히 주관적이고 판단하기 힘들기 때문에 주어진 정보(labeling이 된 supervied방법)으로 찾겠다. 

### 3. 텍스트 분류란?
* **벡터 형태**로 표현된 텍스트를 말그대로 분류하는 방법이다.


* DNN이 비약적으로 성능 향상을 가져오기 전까지는 SVM이 가장 많이 사용되었다.


* 현재는 **딥러닝의 CNN, LSTM을 분류 모델**로 가장 많이 사용한다.
Reference : Yoon Kim, Convolutional Neural Networks for Sentence Classification, 2014

* 실제로 텍스트 분류 작업은 큰 범위에서 대부분의 NLP downstream task를 포함한다.
하나의 예로, Siamese Network라는 걸 통해서 Question-Answering pair를 학습하게 되면, 분류의 기준이 "특정 질문에 맞는 정답을 잘 골랐는가 아닌가(0 / 1)"를 해결하는 문제로 바뀌게 된다.

#### NSMC에서는 긍정1/부정0 이 정해져 있는데, 텍스트가 들어왔을 때, 1/0 분류해 주는게, 이 텍스트 안에 있는  감정이 뭐냐고 분류해주는것과 동일하다. 
#### 어떤 데이터셋을 training시키고, 테스트 데이터셋을 넣었을 때, 1/0으로 텍스트 분류해주는게, 영화 리뷰가 긍정/부정을 판단해주는것과 동일하다.
#### 그래서 NSMC를 감정분석 하기 위해서 text classification을 쓰겠다.
