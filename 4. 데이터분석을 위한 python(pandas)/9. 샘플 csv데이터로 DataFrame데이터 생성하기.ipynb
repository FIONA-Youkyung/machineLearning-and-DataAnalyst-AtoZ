### 3. DataFrame에서 원하는 row만 선택하기
row선택을 배우기 전에, 먼저 dataframe의 slicing에 대해서 생각해보자. 
- dataframe의 경우 기본적으로 [ ]연산자가 column선택에 사용된다. 즉, 
    - train_data[0] : 은 0번째 row를 말하는 것이 아니고 0이라는 column을 말한다. (여기선 0 이라는 column이 없기 때문에 에러가 난다. )
- 그런데! slicing을 할 때는 row레벨로 지원한다. 
    - 예를 들어 train_data[:10] 이라고 하면 0~9번째 row를 반환하는 것이다. 

```python
train_data[7:10]
```

#### row 선택하기
- Series의 경우, [ ]로 row선택이 가능하나, DataFrame의 경우는 기본적으로 column을 선택하도록 설계되있어서 [ ]로만 하면 안되고, 
- .loc, .iloc으로 row선택이 가능하다. 
    - loc : Dataframe 자체의 인덱스 자체를 사용
    - iloc : 0 based index 로 사용
    - 이 두함수는 ','을 사용하여 column선택도 가능하다. 

```python
train_data.head() #0 based index라는 것을 알 수 있다. 이러면 loc과 iloc의 차이가 별로 없기 때문에 학습용으로 인덱스를 다시 바꿔보면, 
```

```python
train_data.index= np.arange(100, 991)
train_data.head(5)
```

#### 1) loc

```python
train_data.loc[986]
```

```python
train_data.loc[[100, 981, 562]]
```

#### 2) iloc

```python
train_data.iloc[0]
```

```python
train_data.iloc[[45, 12, 1]]
```

#### 3) row와 column 동시에 선택하기
- loc, iloc속성을 이용할 때, 콤마를 이용하여 둘 다 명시 가능

```python
train_data.loc[[986, 100, 110, 990], ['Survived', 'Name', 'Sex']] #loc 하고 []앞은 row, []뒤는 column
```

```python
train_data.iloc[[98, 10, 11, 99], [1, 2, 3]] #iloc는  []앞은 row, []뒤는 column인데 이 column도 숫자로 
```
