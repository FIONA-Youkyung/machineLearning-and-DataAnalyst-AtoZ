# 16. Transform 함수의 이해 및 활용하기
group by 에 통계함수 적용하면 원본과 다른 데이터가 생성되는데, Transform 을 사용하면, 원본 데이터에 새로운 속성을 추가하기에 굉장히 유용하다. 
## Transform 함수
- groupby 후 transform함수를 사용하면 원래의 index를 유지한 상태로 통계함수를 적용한다. 
- 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산한다. 
- 따라서 새로 생성된 데이터를 원본 dataframe과 합치기 쉽다. 

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/train.csv')
df.head()
```

```python
df.groupby('Pclass').mean() #이러면 원본 데이터와 다른 dataframe이 생긴다. (row 개수도 class에 따라 달라지고)
```

```python
df.groupby('Pclass').transform(np.mean) #이렇게 transform을 이용하게 되면 원본을 유지하면서, 각 그룹별로 통계함수를 계산한다. 
```

```python
#그리고 원본과 index가 같기 때문에 원본 dataframe에 이 데이터를 추가하기에도 굉장히 용의하다. 
#예를 들어 Pclass를 기준으로 한 age의 mean을 age2 column을 만들어서 원본에 추가하고 싶다면, 
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age'] #이렇게 굉장히 쉽게 추가할 수 있다. 
df.head()
```

```python
#이제 multifle하게 해보면
df.groupby(['Pclass', 'Sex']).transform(np.mean).head()
```

```python
df['Age3'] = df.groupby(['Pclass', 'Sex']).transform(np.mean)['Age']
df.head()
```
