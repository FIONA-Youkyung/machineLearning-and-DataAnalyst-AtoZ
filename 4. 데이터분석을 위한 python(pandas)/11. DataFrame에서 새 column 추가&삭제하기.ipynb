# 11. DataFrame에서 새 column 추가&삭제하기

```python
import pandas as pd
```

```python
train_data = pd.read_csv('./data/train.csv')
train_data.head()
```

## 1. 새 column 추가하기
- [ ]을 사용하여 추가하기
- insert 함수를 사용하여 원하는 위치에 추가하기

```python
#1
train_data['Age_double'] = train_data['Age'] *2 #broading casting
 #이걸 'Age_double'이라는 column을 추가하여 넣고 싶다. 
```

```python
train_data[['Age', 'Age_double']].head()
```

```python
#2
train_data['Age_triple'] = train_data['Age']+train_data['Age_double']
train_data.head()
```

```python
#그런데 위의 결과에서 알 수 있듯이, 새롭게 추가되는 column은 항상 맨 뒤에 추가된다. 
#이를 보완할 때 쓸 수 있는 것이 insert함수!
# train_data.insert(loc, column, value, allow_duplicates=False)

train_data.insert(3, 'Fare10', train_data['Fare']+10) #위치, column이름, data값
train_data.head()
```

## 2. column삭제하기
- drop함수를 사용하여 삭제
    - 리스트를 사용하여 멀티플 삭제 가능
    - train_data.drop(
    labels=None,
    axis=0,
    index=None,
    columns=None,
    level=None,
    inplace=False,
    errors='raise',
)

```python
train_data.drop('Age_triple', axis = 1).head()#axis=0 : 행, axis=1 : 열
```

```python
train_data.head()#그러나 원본에서는 남아있다. 
```

```python
train_data.drop(['Age_double', 'Age_triple'], axis = 1, inplace=True) #그럴 땐, inplace를 이용하면 원본을 수정할 수 있다. 
train_data.head()
```
