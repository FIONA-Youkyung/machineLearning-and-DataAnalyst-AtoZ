우수 고객: 가장 소비를 많이 한 고객
코호트: 특정 기간동안 같은 경험을 한 고객들의 집합 

즉, 코호트 분석이란, 고객의 재구매률을 구매하는 것이 된다. 

### 0. 모듈 불러오기

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from datetime import datetime

%matplotlib inline
```

### 1. 데이터 불러오고 데이터 들여다보는건 지난 시간에 했으니 필요한 전처리 해주기

```python
dtypes = {
    'UnitPrice' : np.float32,
    'CustomerID' : np.int32,
    'Quantity' : np.int32
}

retail = pd.read_csv('/Users/jeonghyeonjeong/for github/머신러닝_데이터분석_dataset/shoppingmall_data/OnlineRetailClean.csv')
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'], infer_datetime_format=True)
retail.head()
```

### 3. 고객 코호트 분석
- 일주일이 지날 때마다 재 구매율이 얼마나 되는지 

사용자 기준으로 최초 구매한 월(month) 연산하기
- Month: 구매월 (일(day)는 무시)
- MonthStarted: 사용자가 최초 구매한 달

##### 1) #먼저 Month 대로 (일은 상관없도록 1로 통일) 새로운 Month라는 column 만듬

```python

def extract_month(date):
    return datetime(date.year,date.month,1)

retail['Month'] = retail['InvoiceDate'].apply(extract_month)
```

```python
retail.head()
```

##### 2) MonthStarted: 사용자가 최초 구매한 달

```python
month_group = retail.groupby('CustomerID')['Month']
```

```python
#가장 빠른 달을 찾아 (np.min)
#행을 추가('MonthStarted')해서 기입 (.transform(np.min))
retail['MonthStarted'] = month_group.transform(np.min)
```

```python
retail.tail()
```

##### 3) 현재 구매 달(['Month'])과, 최초로 구매한 달(['MonthStarted'])을 알기 때문에 그 사이 기간을 계산 하여 새로운 columns을 만들 수 있다. 

```python

retail['MonthPassed'] = (retail['Month'].dt.year - retail['MonthStarted'].dt.year)*12 + \
(retail['Month'].dt.month - retail['MonthStarted'].dt.month)

retail.tail()                         
     
```

##### 4) MonthPassed를 기준으로 고객 카운팅
- 기준이 되는 월과 그 월로부터 지난 기간의 고객의 수를 계산

```python
cohort_group = retail.groupby(['MonthStarted', 'MonthPassed'])
cohort_group.tail()

#unique한 'CustomerID' 개수만 추출해야 한다.
def get_unique_no(x):
    return len(np.unique(x))

cohort_df = cohort_group['CustomerID'].apply(get_unique_no).reset_index()
```

```python
cohort_df.head()
```

```python
#이때 MonthPassed의 값들을 column에 넣고, MonthStarted는 row, CustomerID가 값이 되는 형태로 만들고 싶다면

cohort_df = cohort_df.pivot(index='MonthStarted', columns = 'MonthPassed')
cohort_df.head()
```

```python
#이제 비율로 바꿔야 한다. 
#재구매율이란, MonthStarted 기준으로 1달 지났을때 얼마, 2달 지났을 때 얼마... 이런 것!
#따라서 MonthPassed가 0인 첫 행이 기준이 된다. 
customer_cohort = cohort_df.div(cohort_df.iloc[:, 0], axis=0)*100
customer_cohort = customer_cohort.round(decimals=2)
customer_cohort

```

#### 5) heat map으로 시각화
- seaborn의 heatmap함수로 시각화

```python
xticks = np.arange(0, 13)
yticks = ['2010/12', '2011/01', '2011/02', '2011/03', '2011/04', '2011/05', '2011/06', '2011/07', '2011/08', '2011/09', '2011/10', '2011/11', '2011/12']

plt.figure(figsize = (15, 8))
sns.heatmap(customer_cohort, 
            annot=True, 
            xticklabels=xticks,
            yticklabels=yticks, 
            fmt='.1f')
```
