# 모빌리티에서 풀고 있는 다양한 문제들 - 카쉐어링

카쉐어링 단계 : 차량 예약-> 존 방문-> 차량 이용-> 차량 반납

## 데이터 기반 존 및 차량 운영 전략 수립
- 어떤 위치에 존으르 만들어야 할까?
- 존에 있는 차량 가격을 얼마로 설정해야 할까?
- 어떤 존에 어떤 차량을 넣어야 할까?
- 수요 예측 및 운영 전략 수립(Seasonality가 매우 강함)
- 차량은 언제 구매할 것인가?에 대한 전략

=> 이런 것들이 Operation Research 를 통해 결정되고 있다. 
Operation Research : 수학적 모델링, 통계적 모형, 최적화 기법 등을 활용해 효율적인 의사결정을 돕는 기법
> 수학적 모델링 : 현실 문제를 수학적 언어로 수식화할 수, 최적화 문제로 변환하는 과정

> 최적화 문제 : 특정 집합 위에서 정의도니 실수값, 함수, 정수에 대해 그 값이나 최대/최소가 되는 상태르르 해석하는 문제
